```python
Predictors = ['PctBachDeg25_Over', 'incidenceRate',
       'PctPublicCoverageAlone', 'povertyPercent', 'medIncome',
       'PctEmployed16_Over', 'PctHS25_Over', 'PctPublicCoverage',
       'PctPrivateCoverage', 'PctUnemployed16_Over', 'PctMarriedHouseholds',
       'PctBachDeg18_24', 'PctEmpPrivCoverage', 'PercentMarried', 'PctHS18_24',
       'PctBlack', 'binnedInc']
Target = ['TARGET_deathRate']
X = df[Predictors].values
y = df[Target].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize=True)
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_accuracy = round(lr.score(X_train,y_train) * 100,2)
print('linear regression score is %.4g' %lr_accuracy)
print('Accuracy', 100-(np.mean((np.abs(y_test-lr_pred)/y_test))*100))
    
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(max_depth=15,min_samples_leaf=100)
tree.fit(X_train,y_train)
tree_pred = tree.predict(X_test)
tree_accuracy = round(tree.score(X_train,y_train)*100,2)
print('decision tree score is : %.4g'%tree_accuracy)
print('Accuracy', 100-(np.mean((np.abs(y_test-tree_pred)/y_test))*100))
```

```python
from sklearn.ensemble import RandomForestRegressor
tree = RandomForestRegressor(max_depth=1, criterion='mse', n_estimators=100)
tree.fit(X_train,y_train)
tree_pred = tree.predict(X_test)
tree_accuracy = round(tree.score(X_train,y_train)*100,2)
print('Random forest score is : %.4g'%tree_accuracy)
print('Accuracy', 100-(np.mean((np.abs(y_test-tree_pred)/y_test))*100))
```
